```python
import plotly.io as pio
import pandas as pd
pio.templates
import dash
from dash import dcc
import dash_html_components as html
from dash.dependencies import Input, Output
import plotly.graph_objs as go
import pandas as pd
```

```python
app = dash.Dash()
```

```python
countries = [{'label': c, 'value': c} for c in (covid_1['Country/Region'].unique())]
```

```python
app.layout = html.Div([

    html.Div([
    html.Br(), html.Br(),
    html.H1('Covid 19 Data Dashboard')],
    style={'margin-left': '5%','color':'#808000','width': '50%', 'display': 'inline-block'

    }),

html.Div([
    html.Br(), html.Br()],
    style={'color':'#17202A','width': '30%', 'display': 'inline-block', 'float': 'right'

    }),


# Create scattermapbox map chart
    

    html.Br(), html.Br(),
    html.Div([
        html.Label('Select a Country:'),
        dcc.Dropdown(id='w_countries',
                     multi=False,
                     clearable=True,
                     disabled=False,
                     style={'display': True},
                     value='US',
                     placeholder='Select Countries',
                     options=countries)

         ], style={'width': '10%','margin-left': '45%'}),


    html.Div([
    html.Br(),
    dcc.Graph(id='bar_1'),

        ],style={'margin-left': '1.4%','width': '50%', 'display': 'inline-block'}),


    html.Div([
    html.Br(),
    dcc.Graph(id='pie'),

        ], style={'margin-left': '3.4%', 'width': '40.6%', 'display': 'inline-block', 'float': 'right'}),


    html.Div([
    html.Br(),
    dcc.Graph(id='line_2'),

        ],style={'margin-left': '1.4%','width': '50%', 'display': 'inline-block'}),


    html.Div([
    html.Br(),
    dcc.Graph(id='top_confirmed'),

        ], style={'margin-left': '3.4%', 'width': '40.6%', 'display': 'inline-block', 'float': 'right'}),


   

# Create horizontal bar chart (top deaths)
    html.Div([
    html.Br(),
    dcc.Graph(id='top_deaths'),

        ], style={'margin-left': '3.4%', 'width': '40.6%', 'display': 'inline-block', 'float': 'right'}),


    html.Br(), html.Br(),

    ], style={'background-color': '#e6e6e6'})
```

```python
@app.callback(Output('bar_1', 'figure'),
              [Input('w_countries', 'value')])
def update_graph(w_countries):
# Data for bar
    covid2 = covid_1.groupby(['Country/Region', 'Date'])['New cases'].max().reset_index()

    return {
        'data': [go.Bar(x=covid2[covid2['Country/Region'] == w_countries]['Date'],
                        y=covid2[covid2['Country/Region'] == w_countries]['New cases'],

                        name='Confirmed Cases',
                        marker=dict(
                            color='red'),
                        hoverinfo='text',
                        hovertext=
                        '<b>Country</b>: ' + covid2[covid2['Country/Region'] == w_countries]['Country/Region'].astype(str) + '<br>' +
                        '<b>New Cases</b>: ' + [f'{x:,.0f}' for x in covid2[covid2['Country/Region'] == w_countries]['New cases']] + '<br>' +
                        '<b>Date</b>: ' + covid2[covid2['Country/Region'] == w_countries]['Date'].astype(str) + '<br>'


                        )],


        'layout': go.Layout(
             width=920,
             height=520,
             title={
                'text': 'Daily New Cases : ' + (w_countries),
                'y': 0.93,
                'x': 0.43,
                'xanchor': 'center',
                'yanchor': 'top'},
             titlefont={'family': 'Oswald',
                        'color': 'rgb(0, 0, 0)',
                        'size': 25},

             hovermode='x',

             xaxis=dict(title='<b>Date</b>',
                        color='rgb(0, 0, 0)',
                        showline=True,
                        showgrid=True,
                        showticklabels=True,
                        linecolor='rgb(104, 204, 104)',
                        linewidth=2,
                        ticks='outside',
                        tickfont=dict(
                            family='Arial',
                            size=12,
                            color='rgb(17, 37, 239)'
                        )

                ),

             yaxis=dict(title='<b>New Cases</b>',
                        color='rgb(0, 0, 0)',
                        showline=True,
                        showgrid=True,
                        showticklabels=True,
                        linecolor='rgb(104, 204, 104)',
                        linewidth=2,
                        ticks='outside',
                        tickfont=dict(
                           family='Arial',
                           size=12,
                           color='rgb(17, 37, 239)'
                        )

                ),

             legend=dict(title='',
                         x=0.25,
                         y=1.08,
                         orientation='h',
                         bgcolor='rgba(255, 255, 255, 0)',
                         traceorder="normal",
                         font=dict(
                              family="sans-serif",
                              size=12,
                              color='#000000')),

                 )

    }
```

```python
@app.callback(Output('line_2', 'figure'),
              [Input('w_countries', 'value')])
def update_graph(w_countries):
# Data for line
    covid3 = covid_1.groupby(['Country/Region', 'Date'])[['New cases', 'New deaths', 'New recovered']].max().reset_index()


    return {
        'data': [go.Scatter(
                     x=covid3[covid3['Country/Region'] == w_countries]['Date'],
                     y=covid3[covid3['Country/Region'] == w_countries]['New cases'],
                     name='New cases',

                     mode='lines',
                     marker=dict(color='#9C640C'),

                     hoverinfo='text',
                     hovertext=
                     '<b>Country</b>: ' + covid3[covid3['Country/Region'] == w_countries]['Country/Region'].astype(str) + '<br>'+
                     '<b>New Cases</b>: ' + [f'{x:,.0f}' for x in covid3[covid3['Country/Region'] == w_countries]['New cases']] + '<br>'+
                     '<b>Date</b>: ' + covid3[covid3['Country/Region'] == w_countries]['Date'].astype(str) + '<br>'

                ),
              go.Bar(
                  x=covid3[covid3['Country/Region'] == w_countries]['Date'],
                  y=covid3[covid3['Country/Region'] == w_countries]['New deaths'],
                  name='New deaths',

                  marker=dict(color='red'),

                  hoverinfo='text',
                  hovertext=
                  '<b>Country</b>: ' + covid3[covid3['Country/Region'] == w_countries]['Country/Region'].astype(str) + '<br>' +
                  '<b>New Deaths</b>: ' + [f'{x:,.0f}' for x in covid3[covid3['Country/Region'] == w_countries]['New deaths']] + '<br>' +
                  '<b>Date</b>: ' + covid3[covid3['Country/Region'] == w_countries]['Date'].astype(str) + '<br>'
              ),
             go.Bar(
                  x=covid3[covid3['Country/Region'] == w_countries]['Date'],
                  y=covid3[covid3['Country/Region'] == w_countries]['New recovered'],
                  name='New recovered',

                  marker=dict(color='#008000'),

                  hoverinfo='text',
                  hovertext=
                  '<b>Country</b>: ' + covid3[covid3['Country/Region'] == w_countries]['Country/Region'].astype(str) + '<br>' +
                  '<b>New Recovered</b>: ' + [f'{x:,.0f}' for x in covid3[covid3['Country/Region'] == w_countries]['New recovered']] + '<br>'+
                  '<b>Date</b>: ' + covid3[covid3['Country/Region'] == w_countries]['Date'].astype(str) + '<br>'
              )
        ],


        'layout': go.Layout(
             barmode='stack',
             width=920,
             height=520,
             title={
                'text': 'Daily New Cases : ' + (w_countries),
                'y': 0.93,
                'x': 0.43,
                'xanchor': 'center',
                'yanchor': 'top'},
             titlefont={'family': 'Oswald',
                        'color': 'rgb(0, 0, 0)',
                        'size': 25},

             hovermode='x',

             xaxis=dict(title='<b>Date</b>',
                        color='rgb(0, 0, 0)',
                        showline=True,
                        showgrid=True,
                        showticklabels=True,
                        linecolor='rgb(104, 204, 104)',
                        linewidth=2,
                        ticks='outside',
                        tickfont=dict(
                            family='Arial',
                            size=12,
                            color='rgb(17, 37, 239)'
                        )

                ),

             yaxis=dict(title='<b>New Cases</b>',
                        color='rgb(0, 0, 0)',
                        showline=True,
                        showgrid=True,
                        showticklabels=True,
                        linecolor='rgb(104, 204, 104)',
                        linewidth=2,
                        ticks='outside',
                        tickfont=dict(
                           family='Arial',
                           size=12,
                           color='rgb(17, 37, 239)'
                        )

                ),

             legend=dict(title='',
                         x=0.25,
                         y=1.08,
                         orientation='h',
                         bgcolor='rgba(255, 255, 255, 0)',
                         traceorder="normal",
                         font=dict(
                              family="sans-serif",
                              size=12,
                              color='#000000')),


                 )

    }
```

```python
@app.callback(Output('top_confirmed', 'figure'),
              [Input('w_countries', 'value')])
def update_graph(w_countries):
    top_confirmed = covid_1.groupby(['Country/Region'])[['Confirmed', 'Deaths', 'Recovered', 'Active']].max().sort_values(by=['Confirmed'], ascending=False).nlargest(10, columns=['Confirmed']).reset_index()


    return {
        'data': [go.Bar(x=top_confirmed['Confirmed'],
                            y=top_confirmed['Country/Region'],
                            text=top_confirmed['Confirmed'],
                            texttemplate='%{text:.2s}',
                            textposition='auto',
                            marker=dict(color='#9C640C'),
                            orientation='h',
                            hoverinfo='text',
                            hovertext=
                            '<b>Country</b>: ' + top_confirmed['Country/Region'].astype(str) + '<br>' +
                            '<b>Confirmed Cases</b>: ' + [f'{x:,.0f}' for x in top_confirmed['Confirmed']] + '<br>' +
                            '<b>Deaths</b>: ' + [f'{x:,.0f}' for x in top_confirmed['Deaths']] + '<br>' +
                            '<b>Recovered Cases</b>: ' + [f'{x:,.0f}' for x in top_confirmed['Recovered']] + '<br>' +
                            '<b>Active Cases</b>: ' + [f'{x:,.0f}' for x in top_confirmed['Active']] + '<br>'
                           )],



        'layout': go.Layout(
             width=640,
             height=520,
            
             title={
                'text': 'Top Confirmed Cases: 22-01-2020 t0 27-07-2020 ',
                'y': 0.93,
                'x': 0.43,
                'xanchor': 'center',
                'yanchor': 'top'},
             titlefont={'family': 'Oswald',
                        'color': 'rgb(0, 0, 0)',
                        'size': 25},

             hovermode='closest',
             margin=dict(l=100),

             xaxis=dict(title='<b>Confirmed</b>',

                        color='rgb(0, 0, 0)',
                        showline=True,
                        showgrid=True,
                        showticklabels=True,
                        linecolor='rgb(104, 204, 104)',
                        linewidth=2,
                        ticks='outside',
                        tickfont=dict(
                            family='Arial',
                            size=12,
                            color='rgb(17, 37, 239)'
                        )

                ),

             yaxis=dict(title='<b></b>',
                        autorange='reversed',
                        color='rgb(0, 0, 0)',
                        showline=True,
                        showgrid=False,
                        showticklabels=True,
                        linecolor='rgb(104, 204, 104)',
                        linewidth=2,
                        ticks='outside',
                        tickfont=dict(
                           family='Arial',
                           size=12,
                           color='rgb(17, 37, 239)'
                        )

                )

        )
    }
```

```python
@app.callback(Output('pie', 'figure'),
              [Input('w_countries', 'value')])
def display_content(w_countries):
    covid4 = covid_1.groupby(['Country/Region'])[['Confirmed', 'Deaths', 'Recovered', 'Active']].max().reset_index()
    confirmed = covid4['Confirmed'].sum()
    deaths = covid4['Deaths'].sum()
    recovered = covid4['Recovered'].sum()
    active = covid4['Active'].sum()
    colors = ['#9C640C', '#000000', '#008000', '#00FFFF']

    return {
        'data': [go.Pie(labels=['Confirmed Cases', 'Deaths', 'Recovered Cases', 'Active Cases'],
                        values=[confirmed, deaths, recovered, active],
                        marker=dict(colors=colors),
                        hoverinfo='label+value+percent',
                        textinfo='label+value+percent',
                        textfont=dict(size=13),
                        hole=.7,
                        rotation=45
                        # insidetextorientation='radial',


                        )],

        'layout': go.Layout(
            width=640,
            height=520,
            hovermode='closest',
            title={
                'text': 'Total Global Cases: 22-01-2020 to 27-07-2020',
                'y': 0.96,
                'x': 0.5,
                'xanchor': 'center',
                'yanchor': 'top'},
            titlefont={'family': 'Oswald',
                       'color': 'rgb(0, 0, 0)',
                       'size': 25},
            legend={
                'orientation': 'v',
                'bgcolor': 'rgba(255,255,255,0)',
                'xanchor': 'center', 'x': 0.05, 'y': 1.08},
            ),


        }


```
